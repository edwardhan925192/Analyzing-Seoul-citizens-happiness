```python
import navercrawl
from wordcloud import WordCloud
```

```python
df = navercrawl.news_list(number_of_page=30,sort_how='date')
```

```python
happiness = navercrawl.korean_values_count(df['title'])
```

#WordCloud로 키워드 찾기

```python
wc = WordCloud(background_color='white', font_path = "NanumBarunGothic")
cloud = wc.generate_from_frequencies(dict(happiness))
plt.imshow(cloud)
plt.axis('off')
plt.show()
cloud.to_file('./navercloud.jpg')
```
