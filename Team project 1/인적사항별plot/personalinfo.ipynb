#1.필요한 데이터 불러오기

```python
df = pd.read_csv('/content/drive/MyDrive/data/행복도 데이터/seoulsurvey_newest.csv')
```

```python
df['행복지수(가정생활)'] = df['행복지수(가정생활)'].astype(int)
```

#2. 쓰고싶은 데이터 전처리

```python
degrees = pd.DataFrame(df.groupby(['학력'])['행복지수(가정생활)'].mean())
degrees['학력'] = degrees.reset_index().index
degrees['학력'] = degrees['학력'] +1
```

```python
religion = pd.DataFrame(df.groupby(['종교 '])['행복지수(가정생활)'].mean())
religion['종교'] = religion.reset_index().index
religion['종교'] = religion['종교'] +1
```

```python
jobs = pd.DataFrame(df.groupby(['직업'])['행복지수(가정생활)'].mean())
jobs['직업'] = jobs.reset_index().index
jobs['직업'] = jobs['직업'] +1
```

#3.그림그리기

```python
from matplotlib import figure
fig = plt.figure(figsize = (10, 5))
plt.subplots_adjust(left=0.1, right=1.3)
ax1 = fig.add_subplot(1,3,1)	
ax2 = fig.add_subplot(1,3,2)	
ax3 = fig.add_subplot(1,3,3)	

sb.lineplot(degrees, x = '학력', y = '행복지수(가정생활)',marker='o' ,ax = ax1)
sb.lineplot(religion, x = '종교', y = '행복지수(가정생활)',marker='o' ,color='r',ax = ax2)
sb.lineplot(jobs, x = '직업', y = '행복지수(가정생활)',marker='o' ,color='g',ax = ax3)
```
