#1. Columns 이름 바꾸기
#2. 2018 ~ 2021 데이터 합치기

```python
data_path = glob.glob('/content/drive/MyDrive/data/공통 항목 data/*.csv')

```

```python
bupyo = pd.read_excel('./공통부표원.xlsx')
```

```python
bupyo.drop(20,axis = 0 ,inplace = True)
```

```python
bupyo.drop(0,axis = 0 ,inplace = True)
```

```python
col_name = bupyo['Unnamed: 0'].tolist()
```

#년도 column 생성하고 각 년도 넣기 

```python
df_merged2 = pd.read_csv(data_path[0])
df_merged2.columns = col_name
df_merged2['year'] = 2018
count = 2019
print(len(df_merged2.columns))
for year in data_path:
  temp = pd.read_csv(year)
  temp.columns = col_name
  temp['year'] = count
  count += 1   
  df_merged2 = pd.concat((df_merged2,temp),axis = 0) 
```

```python
df_merged2.to_csv('merged_survey.csv',index = False)
```
